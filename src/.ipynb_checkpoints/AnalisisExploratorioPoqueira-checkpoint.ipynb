```python
# Función para convertir duraciones a horas
def convert_duration_to_hours(duration):
    if isinstance(duration, str) and duration.strip():  # Comprobar que es un string no vacío
        parts = duration.split(' ')
        # Comprobar que hay al menos 2 partes (días y tiempo)
        if len(parts) >= 2:
            days = int(parts[0])  # Obtener días
            time_parts = parts[2].split(':')  # Obtener horas:minutos:segundos
            # Comprobar que hay exactamente 3 partes en time_parts
            if len(time_parts) == 3:
                hours = (days * 24 + int(time_parts[0]) + 
                         int(time_parts[1]) / 60 + 
                         int(time_parts[2]) / 3600)
                return hours
            else:
                return np.nan  # Si no hay 3 partes en time_parts, devolver NaN
    return np.nan  # Si no es un string o está vacío, devolver NaN

# Aplicar la conversión a las columnas que contienen duración
data['total_time_POQ'] = data['total_time_POQ'].apply(convert_duration_to_hours)

# Comprobar la conversión
print("\nDespués de la conversión:")
print(data['total_time_POQ'].head())
```
